```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df =pd.read_csv('/content/drive/MyDrive/ML projects datasets/Handling Missing Values practice/credit_data.csv')
```

#This Dataset is highly unbalanced

0--> Normal

1--> fraud

Separating the data for  analysis

```python
legit =df[df.Class == 0]
fraud = df[df.Class == 1]
```

```python
legit.shape , fraud.shape
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare Values for both Transactions

df.groupby('Class').mean()
```

#Dealing unbalance data

Under Sampling

Bulid a sample dataset containing similar distribution of normal and fraud transactions

```python
df['Class'].value_counts()
```

```python
legit_samples = legit.sample(n=492)
```

Concatnating two dataFrames

```python
new_data =pd.concat([legit_samples,fraud],axis =0)
```

```python
new_data.shape
```

```python
new_data.head()
```

```python
new_data.tail()
```

```python
new_data['Class'].value_counts()
```

```python
new_data.groupby('Class').mean()
```

#Spliting data into features and targets

```python
x =new_data.drop(columns='Class',axis=1)
y =new_data['Class']
```

```python
x.columns
```

```python
y.head()
```

#Spliting data into train and Testing


```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2,  stratify=y,random_state=2)
```

```python
y.shape , y_train.shape , y_test.shape
```

#Model Training

LogisticRegression model

```python
lr_model= LogisticRegression()
```

```python
lr_model.fit(x_train,y_train)
```

#Evalution

```python
x_train_prediction = lr_model.predict(x_train)
```

```python
acc_train = accuracy_score(x_train_prediction,y_train)
```

```python
print("Accuracy score:", acc_train)
```

#Model Testing

```python
testing_prediction = lr_model.predict(x_test)
```

```python
acc_test =accuracy_score(testing_prediction,y_test)
```

```python
print("Accuracy score:", acc_test)
```
